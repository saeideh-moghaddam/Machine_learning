```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from scipy.stats import linregress
```

```python
X = np.random.uniform(0, 12, 100)
Y = np.random.normal(X, 1, 100)

plt.scatter(X, Y)
plt.show()
```

```python
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

print(X.shape)
print(Y.shape)
```

```python
m_real = np.matmul(inv(np.matmul(X.T, X)), np.matmul(X.T, Y))
m_real = m_real[0][0]
m_real

m_rand = np.random.normal(m_real, 0.5, 2)

print(m_real)
m_rand
```

```python
plt.scatter(X, Y, c='blue')
plt.plot(X, m_real*X, c='red')
plt.plot(X, X*m_rand[0], c='Turquoise')
plt.plot(X, X*m_rand[1], c='Maroon')
plt.show()
```

```python
X = X.reshape(-1)
Y = Y.reshape(-1)
slope, intercept, r, p, se = linregress(X, Y)
y_pred = intercept + (slope*X)
plt.scatter(X, Y, c='blue')
plt.plot(X, y_pred, c='red')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X, Y, c='blue')
plt.plot(X, y_pred, c='red')

plt.subplot(1, 2, 2)
plt.scatter(X, Y, c='blue')
plt.plot(X, m_real*X, c='red')
plt.plot(X, X*m_rand[0], c='Turquoise')
plt.plot(X, X*m_rand[1], c='Maroon')
plt.show()
```
